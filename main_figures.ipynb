```python
import numpy as np
import matplotlib.pyplot as plt

results_vgg16_summary = np.loadtxt("./results_vgg16_summary.csv",delimiter=",")
results_vgg16_base = np.loadtxt("./results_vgg16_base.csv",delimiter=",")
results_CNN_summary = np.loadtxt("./results_CNN_summary.csv",delimiter=",")
results_CNN_base = np.loadtxt("./results_CNN_base.csv",delimiter=",")
```

```python

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,7),dpi=120)
fig.suptitle('VGG16 with Convolution and a Summary Layer')
ax1.plot(results_vgg16_summary[:,0], results_vgg16_summary[:,1])
ax1.plot(results_vgg16_base[:,0], results_vgg16_base[:,1])
ax1.set_title("Runtime")
ax2.plot(results_vgg16_summary[:,0], results_vgg16_summary[:,3])
ax2.plot(results_vgg16_base[:,0], results_vgg16_base[:,3])
ax2.set_title("Accuracy on test")
ax1.set_xlabel('Epochs', fontsize=16)
ax2.set_xlabel('Epochs', fontsize=16)

```

```python

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,7),dpi=120)
fig.suptitle('CNN with Convolution and a Summary Layer')
ax1.plot(results_CNN_summary[:,0], results_CNN_summary[:,1])
ax1.plot(results_CNN_base[:,0], results_CNN_base[:,1])
ax1.set_title("Runtime")
ax2.plot(results_CNN_summary[:,0], results_CNN_summary[:,3])
ax2.plot(results_CNN_base[:,0], results_CNN_base[:,3])
ax2.set_title("Accuracy on test")
ax1.set_xlabel('Epochs', fontsize=16)
ax2.set_xlabel('Epochs', fontsize=16)

```
